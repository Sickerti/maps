```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd 


```

```python
data = pd.read_excel('Beispiele.xlsx')
data.head()
```

```python
m = folium.Map(location=[53, 10], zoom_start=5)
marker_cluster = MarkerCluster().add_to(m)
```

```python
for i in range(0,len(data)):
    folium.Marker(
        location=[data.iloc[i]['lat'], data.iloc[i]['lon']], popup=data.iloc[i]['Stadt'],
        ).add_to(marker_cluster)
m
```
