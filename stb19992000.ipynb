```python
import folium
from folium import plugins
from folium.plugins import HeatMap

import pandas as pd
df = pd.read_excel('stb19992000.xlsx')
```

```python
df = pd.read_excel('stb19992000.xlsx')

from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="stbapp")

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
df['location'] = df['Ort'].apply(geocode)

df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)

df.to_excel('stb19992000b.xlsx')
```

```python
data = pd.read_excel('stb19992000b.xlsx')
data['loc'] = (data.point.str.split(',', expand=True))[1]
data['point'] = (data.point.str.split(',', expand=True))[0]
data['point']

data.head()
```

```python
m = folium.Map(location=[53, 10], zoom_start=5,)

data['point'] = data['point'].astype(float)
data['loc'] = data['loc'].astype(float)

```

```python
heat_df = data[data['StbSignatur']=='2007'] # Reducing data size so it runs faster
heat_df = heat_df[['point', 'loc']]
heat_df = heat_df.dropna(axis=0, subset=['point','loc'])
```

```python
heat_data = [[row['point'],row['loc']] for index, row in heat_df.iterrows()]
```

```python
HeatMap(heat_data).add_to(m)
m
```
