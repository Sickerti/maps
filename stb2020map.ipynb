```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd 
```

```python
data = pd.read_excel('stb20202.xlsx')
data.head()
```

```python
data['loc'] = (data.point.str.split(',', expand=True))[1]
data['point'] = (data.point.str.split(',', expand=True))[0]
data.head()
```

```python
m = folium.Map(location=[53, 10], zoom_start=5)
marker_cluster = MarkerCluster().add_to(m)
```

```python
for i in range(0,len(data)):
    folium.Marker(
        location=[data.iloc[i]['point'], data.iloc[i]['loc']], popup=data.iloc[i]['Ort'],
        ).add_to(marker_cluster)

        
```
